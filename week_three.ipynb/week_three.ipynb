```python
import numpy as np
np.zeros(shape=(4,5)) #shape parameter is given in form of **tuple** data type
```

```python
import matplotlib.pyplot as plt
import PIL 
import urllib
```

```python
#define url of an image we want to load 
# This can also be a path to a local file
image_url= "https://img.freepik.com/free-psd/vibrant-bonsai-tree-decorative-pot-transparent-background_84443-28899.jpg?semt=ais_hybrid"
```

```python
image_data = urllib.request.urlopen(image_url)
image = PIL.Image.open(image_data)
np_image = np.array(image)
```

```python
plt.imshow(np_image)
```
