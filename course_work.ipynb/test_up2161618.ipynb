## Question 1  

```python
# This creates a dimension for the amount of rows and coloums for the data set
rows = 5
coloums = 5

# The numpy array is not created random for each one of the rows and coloums.
pandadata = np.random.rand(rows,coloums)

# This is used to label each row and coloums 
rlabel =  ['A1','B1','C1','D1','E1']
clabel = ['a1','a2','a3','a4','a5']

# The dataframe is now created as a part of pandas using the data from...
#...the numpy array to sort it out into the right row and coloum.
dataframe = pd.DataFrame(pandadata,index= clabel, columns= rlabel)

# This prints out the data frame the way it is, the ''\n'' adds space
print("Data frame:")
print(f"{dataframe}\n")

# This prints out the summary for the dataframe
print("Summary:")
print(f"{dataframe.info()}\n")

# This is used to print out the shape of the dataframe
print("Shape:")
print(f"{dataframe.shape}\n")

# This part of the code is used to print the statistics of the dataframe
print("Statistics:")
print(f"{dataframe.describe()}\n")
```

```python
iris =  sns.load_dataset('iris')

print(iris)

plt.figure(figsize=(6,10))
iris.boxplot()
plt.title("Box plot for each")
plt.xlabel("length and width")
plt.ylabel("Range")

plt.grid(True)
plt.tight_layout()
plt.show
```
